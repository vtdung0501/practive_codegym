### Vẽ Tam giác lồng nhau </p>
cho 2 biến a và b là số nguyên 
Hãy vẽ a hình tam giác đều theo hình sau , với hình trong cùng có kích thước cạnh là b , các hình tiếp theo có kích thước tăng dần 20%


```python
import turtle as tl
#a = int(input("Nhập số lượng hình:"))
#b = int(input("Nhập độ dài cạnh:"))
a=10
b=100
pen = tl.Turtle()
def draw_triangle(b):
    for i in range(0,3):
        pen.left(120)
        pen.forward(b)
for i in range(0,a):
    b = b*1.2 +i
    pen.penup()
    pen.goto(b/2,0)
    pen.pendown()
    draw_triangle(b)
tl.done()
```

### Tính điểm trung bình
hãy viết chương trình nhập điểm của một học sinh có 3 điểm toán lý hóa và hãy tính điểm trung bình của học sinh đó xếp loại gì .</p>
cho biết mức điểm : trung bình: 5 <= tb <= 6 , Khá : 6 < k < 8 ,8 <= g</p>
            hãy xuất ra vd:</p>
                toán : 8 , lý : 8 , Hóa : 8 </p>
                            hiện lên màn hình :</p> 
                            tb : False</p>
                                                k : False</p>
                                                g : True </p>

```python
diem_toan = float(input("Mời nhập điểm Toán:"))
diem_ly = float(input("Mời nhập điểm Lý:"))
diem_hoa = float(input("Mời nhập điểm Hóa:"))
diem_tb = (diem_toan+diem_ly+diem_hoa)/3
if(5<=diem_tb<=6):
    print("tb :", False)
elif(6<diem_tb<8):
    print("k :",False)
elif(diem_tb>=8):
    print("g :",True)
```

### Tính diện tích
Có một mảnh đất như hình dưới , cho biết các cạnh tương ứng và các hình như dưới . 
Các bạn hãy vận dụng kiến thức để tính chu vi,diện tích từng hình và diện tích tổng của mảnh đất.

```python
import math
#Chu vi hình chữ nhật
perimeter_rect = 2*(100+50)
print("Chu vi hình chữ nhật là:",perimeter_rect)
#Chu vi hình vuông
perimeter_square = 4*10
print("Chu vi hình vuông là:",perimeter_square)
#Chu vi hình tam giác
perimeter_triangle = 30+50+10 + math.sqrt(30**2 + (50+10)**2)
print("Chu vi hình tam giác là:",perimeter_triangle)
#Diện tích hình chữ nhật
area_rect = 50*100
print("Diện tích hình chữ nhật là:",area_rect)
# Diện tích hình vuông
area_square = 10**2
print("Diện tích hình vuông là:",area_square)
#Diện tích tam giác
area_triangle = ((50+10)*30)/2
print("Diện tích hình tam giác là:",area_triangle)
sum_area = area_rect+area_square+area_triangle
print("Tổng diện tích mảnh đất là:",sum_area)
```

#### thực hành vẽ mặt cười

```python
import turtle as tl
pen = tl.Turtle()
pen.pensize(5)
pen.color("brown")
#vẽ mặt
pen.fillcolor("yellow")
pen.begin_fill()
face_size = 100
pen.circle(face_size)
pen.end_fill()
#vẽ mắt
pen.penup()
pen.goto(50,100)
pen.pendown()
eye_size = 20
pen.fillcolor("green")
pen.begin_fill()
pen.circle(eye_size)
pen.penup()
pen.goto(-50,100)
pen.pendown()
pen.circle(eye_size)
pen.end_fill()
#vẽ mũi
pen.penup()
pen.goto(0,80)
pen.pendown()
nose_size=15
pen.fillcolor("violet")
pen.begin_fill()
pen.circle(nose_size,steps=3)
pen.end_fill()
#vẽ miệng
pen.penup()
pen.goto(-30,70)
pen.pendown()
pen.right(90)
mouth_size = 30
pen.fillcolor("pink")
pen.begin_fill()
pen.circle(mouth_size,180)
pen.left(90)
pen.forward(60)
pen.end_fill()
tl.done()
```
